```python
popular_item = menu_detail.groupby('ITEM_NAME').agg({'QUANTITY': 'count','REVENUE': 'sum'}).reset_index()
popular_item.head()
```

```python
popular_item = popular_item.rename(columns=({'REVENUE':'Total_revenue'}))
popular_item.head()
```

Menyu analizi və optimallaşdırma  ↓ 

```python
fig,ax = plt.subplots(figsize=(12,6))
sns.scatterplot(x='QUANTITY', y='Total_revenue', data=popular_item, s=100, alpha=0.7, edgecolor='black')
q_med = popular_item['QUANTITY'].median()
r_med = popular_item['Total_revenue'].median()
top_revenue = popular_item.nlargest(5, 'Total_revenue')
top_quantity = popular_item.nlargest(5, 'QUANTITY')

top_items = pd.concat([top_revenue, top_quantity]).drop_duplicates()
xmin, xmax = ax.get_xlim()
ymin, ymax = ax.get_ylim()
ax.axvspan(xmin, q_med, ymin=0, ymax=1, facecolor="lightgrey", alpha=0.15)   
ax.axvspan(q_med, xmax, ymin=0, ymax=1, facecolor="lightyellow", alpha=0.12)
ax.axhspan(ymin, r_med, xmin=0, xmax=1, facecolor="lightgrey", alpha=0.08)   
ax.axhspan(r_med, ymax, xmin=0, xmax=1, facecolor="lightgreen", alpha=0.06)
plt.text((xmin + q_med) / 2, (r_med + ymax) / 2, "Potential", ha="center", va="center", fontsize=12, weight="bold")
plt.text((q_med + xmax) / 2, (r_med + ymax) / 2, "Stars", ha="center", va="center", fontsize=12, weight="bold")
plt.text((xmin + q_med) / 2, (ymin + r_med) / 2, "Underperformers", ha="center", va="center", fontsize=12, weight="bold")
plt.text((q_med + xmax) / 2, (ymin + r_med) / 2, "Volume Drivers", ha="center", va="center", fontsize=12, weight="bold")


plt.axvline(q_med, color='red', linestyle='--')
plt.axhline(r_med, color='blue', linestyle='--')
plt.fill_betweenx([0, popular_item['Total_revenue'].max()],
                  0, popular_item['QUANTITY'].median(),
                  color='lightgrey', alpha=0.2)
plt.fill_betweenx([0, popular_item['Total_revenue'].max()],
                  popular_item['QUANTITY'].median(), popular_item['QUANTITY'].max(),
                  color='lightyellow', alpha=0.2)
for _, row in top_items.iterrows():
    plt.text(
        row['QUANTITY'],
        row['Total_revenue'],
        row['ITEM_NAME'],
        fontsize=9
    )

           
plt.title('BCG Matrix: Menu Analysis (Popularity vs Profitability)')
plt.xlabel('Popularity (Quantity Sold)')
plt.ylabel('Total Profit')

plt.tight_layout()
plt.show()
```

"Underperformers" kateqoriyasına daxil olan məhsullar həm satış həcmi, həm də ümumi mənfəət baxımından aşağı nəticə göstərir (populyarlıq ~5–15, profit isə ~40–100 aralığında). RFM-seqmentasiya ↓ 

```python
now = pd.Timestamp.now(tz=order_table['ORDER_DATE'].dt.tz)
order_table['Recency'] = (now - order_table['ORDER_DATE']).dt.days
order_table
```

```python
order_table['Frequency'] = order_table.groupby('CUSTOMER_ID')['ORDER_ID'].count()
```

```python
order_table['Monetary'] = order_table['TOTAL_PRICE']
```

```python
recency_scores = [1,2,3]
frequency_scores = [1, 2, 3]
monetary_scores = [1, 2, 3]
```

```python
order_table['Recency_score']  = pd.cut(order_table['Recency'],bins = 3,labels = recency_scores)
order_table['Frequency_score'] = pd.cut(order_table['Frequency'], bins = 3,labels=frequency_scores)
order_table['Monetary_score'] = pd.cut(order_table['Monetary'], bins = 3,labels = monetary_scores)
```

```python
cols = ['Recency_score','Frequency_score','Monetary_score']
order_table[cols] = order_table[cols].apply(pd.to_numeric, errors='coerce').fillna(0).astype(int)
```

```python
order_table['RFM_score'] = order_table['Recency_score'] + order_table['Frequency_score'] + order_table['Monetary_score']
```

```python
order_table.head()
```

```python
order_table['RFM_segment'] = order_table['Recency_score'].astype(str) + '-' +\
    order_table['Frequency_score'].astype(str) + '-' +\
        order_table['Monetary_score'].astype(str)
```
